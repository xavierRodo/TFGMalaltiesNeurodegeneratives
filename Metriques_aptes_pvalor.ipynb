```python
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)
pss_p=pd.read_csv('pss_p.csv')
pss_p=pss_p.rename(columns={'3':'SC'})
pss_p[['SC1','SC2','SC3','C1C2','C1C3','C2C3']]=pss_p.SC.str.split(',',expand=True)
pss_p=pss_p.drop(columns=['SC'])
pss_p['SC1']=pss_p['SC1'].str[1:]
pss_p['C2C3']=pss_p['C2C3'].str[:-1]
pss_p = pss_p[pss_p["SC1"].str.contains("nan") == False]
pss_p = pss_p[pss_p["SC2"].str.contains("nan") == False]
pss_p = pss_p[pss_p["SC3"].str.contains("nan") == False]
pss_p = pss_p[pss_p["C1C2"].str.contains("nan") == False]
pss_p = pss_p[pss_p["C2C3"].str.contains("nan") == False]
pss_p = pss_p[pss_p["C1C3"].str.contains("nan") == False]
pss_p.drop(columns=['2','SC1','SC2','SC3','C1C2','C1C3','C2C3']).groupby(['0','1']).count().sort_values(by='Unnamed: 0', ascending=False)
```
